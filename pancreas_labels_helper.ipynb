```python
import dicom2nifti
import nibabel as nib
import nilearn as nil
import numpy as np
import scipy.ndimage as ndi
import matplotlib.pyplot as plt
import os

TASK07 = 'monai/prep/Task07_Pancreas/imagesTr'
NIICONV = "niicubes"
```

```python
directory = NIICONV
inputdirectory = NIICONV
images_list = os.listdir(directory)
selected = images_list[np.random.randint(0, len(images_list))]
print(images_list)
print(selected)
```

## Visualization

```python
pancreas_in = nib.load("{0:s}/{1:s}".format(inputdirectory, selected))
pancreas_in_data = pancreas_in.get_fdata()
print(f"Input: {pancreas_in.get_fdata().shape}")
print(pancreas_in.header)
```

```python
pancreas_vol = nib.load("{0:s}/{1:s}".format(directory, selected))
pancreas_vol_data = pancreas_vol.get_fdata()
print(f"Output: {pancreas_vol.get_fdata().shape}")
print(pancreas_vol.header)
```

```python
fig_rows = 4
fig_cols = 4
n_subplots = fig_rows * fig_cols
n_slice = pancreas_in_data.shape[2]
step_size = n_slice // n_subplots
plot_range = n_subplots * step_size
start_stop = int((n_slice - plot_range) / 2)

fig, axs = plt.subplots(fig_rows, fig_cols, figsize=[10, 10])

for idx, img in enumerate(range(start_stop, plot_range, step_size)):
    axs.flat[idx].imshow(pancreas_in_data[:, :, img], cmap='gray')
    axs.flat[idx].axis('off')
        
plt.tight_layout()
plt.show()
```

```python
fig_rows = 4
fig_cols = 4
n_subplots = fig_rows * fig_cols
n_slice = pancreas_vol_data.shape[2]
step_size = n_slice // n_subplots
plot_range = n_subplots * step_size
start_stop = int((n_slice - plot_range) / 2)

fig, axs = plt.subplots(fig_rows, fig_cols, figsize=[10, 10])

for idx, img in enumerate(range(start_stop, plot_range, step_size)):
    axs.flat[idx].imshow(pancreas_vol_data[:, :, img])
    axs.flat[idx].axis('off')
        
plt.tight_layout()
plt.show()
```

```python
fig_rows = 1
fig_cols = 1
n_slice = pancreas_vol_data.shape[2]
fig, axs = plt.subplots(1, 1, figsize=[10, 10])
SLICEOUTPUT = f"output/slice_output.png"
SLICEINPUT = f"output/slice_input.png"
cube = []

for idx, img in enumerate(range(0, pancreas_vol_data.shape[2], 1)):
    axs.imshow(pancreas_vol_data[:, :, img])
    plt.tight_layout()
    plt.savefig(SLICEOUTPUT)
    axs.imshow(pancreas_in_data[:, :, img], cmap='gray')
    plt.tight_layout()
    plt.savefig(SLICEINPUT)

    try:
        from PIL import Image
    except ImportError:
        import Image

    background = Image.open(SLICEOUTPUT)
    overlay = Image.open(SLICEINPUT)

    background = background.convert("RGBA")
    overlay = overlay.convert("RGBA")

    new_img = Image.blend(background, overlay, 0.5)
    cubefile = f"output/output_{selected}_{img}.png"
    new_img.save(cubefile,"PNG")

    print(cubefile)

concatenated = Image.fromarray(
  np.concatenate(
    [np.array(Image.open(x).resize((96,96))) for x in cube],
    axis=1
  )
)
```
